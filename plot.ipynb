```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('./cleaned60.csv', delimiter=",")

# Ensure the column names match your CSV file
# Convert VALUEDATE to datetime and PVALUE to numeric
df['VALUEDATE'] = pd.to_datetime(df['VALUEDATE'], errors='coerce')
df['PVALUE'] = pd.to_numeric(df['PVALUE'], errors='coerce')

# Drop rows with NaN values in either column
df = df.dropna(subset=['VALUEDATE', 'PVALUE'])

# Set VALUEDATE as the index
df.set_index('VALUEDATE', inplace=True)

# Resample data to get one sample per day (or another interval)
# Here, '1D' means one sample per day. You can adjust this to '1H' for hourly, '1T' for minutely, etc.
df_resampled = df.resample('1H').mean()  # Taking the mean value for each hour

# Drop NaN values that might have resulted from the resampling process
df_resampled = df_resampled.dropna()

# Extract the relevant data
time_data = df_resampled.index
energy_data = df_resampled['PVALUE']

# Plot the histogram with time-based bins
plt.figure(figsize=(14, 7))
plt.hist(time_data, bins=50, weights=energy_data, color='blue', edgecolor='black')
plt.title('Energy Measurements Over Time')
plt.xlabel('Time')
plt.ylabel('Power (PVALUE)')
plt.grid(True)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Save the plot as a PNG file
plt.savefig('histogram60.png')

# Show the plot
plt.show()


# Load the CSV file
df = pd.read_csv('./cleaned900.csv', delimiter=",")

# Ensure the column names match your CSV file
# Convert VALUEDATE to datetime and PVALUE to numeric
df['VALUEDATE'] = pd.to_datetime(df['VALUEDATE'], errors='coerce')
df['PVALUE'] = pd.to_numeric(df['PVALUE'], errors='coerce')

# Drop rows with NaN values in either column
df = df.dropna(subset=['VALUEDATE', 'PVALUE'])

# Set VALUEDATE as the index
df.set_index('VALUEDATE', inplace=True)

# Resample data to get one sample per day (or another interval)
# Here, '1D' means one sample per day. You can adjust this to '1H' for hourly, '1T' for minutely, etc.
df_resampled = df.resample('1H').mean()  # Taking the mean value for each hour

# Drop NaN values that might have resulted from the resampling process
df_resampled = df_resampled.dropna()

# Extract the relevant data
time_data = df_resampled.index
energy_data = df_resampled['PVALUE']

# Plot the histogram with time-based bins
plt.figure(figsize=(14, 7))
plt.hist(time_data, bins=50, weights=energy_data, color='green', edgecolor='black')
plt.title('Energy Measurements Over Time')
plt.xlabel('Time')
plt.ylabel('Power (PVALUE)')
plt.grid(True)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Save the plot as a PNG file
plt.savefig('histogram900.png')

# Show the plot
plt.show()
```
